GENERACIÓN ENTORNO DOCKER
---

Una vez tenemos la API creada vamos a generar un entorno Docker donde poder trabajar y poder ejecutar todo correctamente salvando las dependencias de sistemas operativos, librerías, entornos, etc.

Para la creación del entorno DOCKER solo necesitamos dos archivos adiccionales a los que ya tenemos. 

- Dockerfile: Archivo muy parecido al txt, que contiene la configuración de la imagen del contenedor. 

- Requirements.txt: Archivo txt que contiene las librerias necesarias para la ejecución de nuestra API asi como la ejecución de nuestro modelo. 

#### 1.) Creación de Dockerfile:

Para la creación del Dockerfile vamos a usar Visual Studio Code. Ya que el propio editor sabe identificarte el tipo de archivo. Una vez creado. Deberemos guardarlo como **Dockerfile** no valen archivos llamados como *(1.Dockerfile, .Dockerfile, DOCKERFILE o dockerfile).* 

A continuación comentamos la configuración de nuestro Dockerfile: 

- `FROM Ubuntu`: Escogemos la distribución base de Linux, si bien es cierto que aqui se pueden escoger varias opciones `python:versión que necesitas`, `alpine`, incluso distribuciones como centos o windows tambien son valias. Aqui va mas en función de como trabajan. Nosotros probamos alpine y python y no nos terminaron de convencer. Nos decantamos por Ubuntu al final.

- `ADD . /accidentes_api`. Incluimos la carpeta donde localizamos el modelo y la API.

- `WORKDIR /accidentes_api`. Incluimos como directorio de trabajo la carpeta donde localizamos el modelo y la API.

- `RUN apt-get update && apt-get install python3.6 -y && apt-get install python3-pip -y
   RUN apt-get install vim -y
   RUN pip3 install -r requirements.txt`. Este comado lo incluimos para instalar todas las carpetas necesarias en requirements.txt. En los siguientes pasos de este tutorial mostramos como hemos creado la carpeta requirements.txt. 
   
- `CMD ["/bin/bash"]` Este comando se encarga de pasar valores por defecto a un contenedor. Entre estos valores se pueden pasar ejecutables. Como hemos escogido Linux, el comando CMD será `/bin/bash`.
